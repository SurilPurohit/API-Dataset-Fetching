```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=a91e384aa21a2e5a13c973706377cf4e&include_video=false&language=en-US&page=1&sort_by=popularity.desc')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['adult', 'id', 'original_language', 'overview', 'popularity', 'release_date', 'title', 'vote_average', 'vote_count']]
```

```python
temp_df
```

```python
temp_df.set_index('id', inplace=True)
```

```python
temp_df
```

```python
df = pd.DataFrame()
for i in range(1,501):
    response = requests.get(f'https://api.themoviedb.org/3/discover/movie?api_key=a91e384aa21a2e5a13c973706377cf4e&include_video=false&language=en-US&page={i}&sort_by=popularity.desc')
    temp_df = pd.DataFrame(response.json()['results'])
    df = df._append(temp_df,ignore_index=True)
```
